```python
import pandas as pd
average_magnitude = {}
variance = []
for tx in tqdm(range(len(parameters['device_indices'])), disable=True):
        for rx in range(len(parameters['device_indices'])):
            y = []
            for i in range(parameters['num_iter']):
                y.append(np.average(np.abs(dataset[i][tx][rx])))
            # average_magnitude[f'Tx{tx+1} Rx{rx+1}'] = y
            
            average_magnitude[f'Tx{tx+1} Rx{rx+1}'] = y

            variance.append(np.std(y))
```

```python
df_average_magnitude = pd.DataFrame(average_magnitude)
df_average_magnitude
```

```python
df_average_magnitude = df_average_magnitude.apply((lambda x: actual_power(x, parameters['receiver_gain'])))
df_average_magnitude
```

```python
import numpy as np 
def friis_transmission_equation(Pt, Gr, d, freq):
    lamda = 3e8/freq
    Gt = Gr
    Pr = Pt + Gt + Gr + 20*np.log10((lamda/(4*np.pi*d)))

    print(f"Pr = {Pt} + {Gt} + {Gr} + 20*log(10, {lamda:.2f}/(4*pi*{d}))")
    print(f'Pr = {Pr:.2f} dB')

friis_transmission_equation(7, 6.6, 1.5, 2.3e9)
```

```python
def actual_power(MRx, gain):
    PRx = 7 + 20 * np.log10(MRx/1000) - gain

    # print(f'PRx = 7 + 20*log(10, ({MRx}/1000)) - {gain}')
    print(f'PRx = {PRx:.2f} dB')

actual_power(811, 30)
```
